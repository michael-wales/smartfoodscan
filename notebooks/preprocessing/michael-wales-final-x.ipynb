```python
import pandas as pd

# Load a random 0.1% of the data
df = pd.read_csv("../raw_data/openfoodfacts_products.csv",
                 skiprows=lambda x: x > 0 and (x % 1_000 != 0),
                 on_bad_lines='skip',
                 dtype=str,
                 sep='\t')

df.shape
```

```python
pd.set_option("display.max_rows", None)  # Show all rows
pd.set_option("display.max_columns", None)  # Show all columns

df.head()
```

```python
columns_to_keep = [
    "created_datetime", "labels_en", "ingredients_text", "additives_en",
    "nutriscore_score", "nutriscore_grade", "nova_group", "energy-kcal_100g",
    "fat_100g", "saturated-fat_100g", "trans-fat_100g", "cholesterol_100g",
    "carbohydrates_100g", "sugars_100g", "fiber_100g", "proteins_100g",
    "sodium_100g", "calcium_100g", "iron_100g"
]

filtered_chunks = []

# Read the file in chunks
chunk_size = 100000

for chunk in pd.read_csv("../raw_data/openfoodfacts_products.csv",
                         usecols=["countries_en"] + columns_to_keep,
                         chunksize=chunk_size,
                         sep="\t"):

    chunk_filtered = chunk[chunk["countries_en"].isin(["United States", "Canada"])]

    filtered_chunks.append(chunk_filtered)

# Concatenate all filtered chunks into a single DataFrame
filtered_df = pd.concat(filtered_chunks, ignore_index=True)

filtered_df.drop(columns=["countries_en"], inplace=True)

filtered_df.info()
```

```python
filtered_df["created_datetime"] = pd.to_datetime(filtered_df["created_datetime"], format="%Y-%m-%dT%H:%M:%SZ", errors="coerce")
filtered_df = filtered_df[filtered_df["created_datetime"].dt.year >= 2017]
filtered_df.reset_index(drop=True, inplace=True)
print(filtered_df["created_datetime"].min())
print(filtered_df.info())
```

```python
filtered_df.drop(columns=['created_datetime'], inplace=True)
```

```python
columns_to_check = [
    "energy-kcal_100g", "fat_100g", "saturated-fat_100g", "trans-fat_100g",
    "cholesterol_100g", "sodium_100g", "carbohydrates_100g", "fiber_100g",
    "sugars_100g", "proteins_100g", "calcium_100g", "iron_100g"
]

# Drop rows where any of these columns have NaN values
filtered_df = filtered_df.dropna(subset=columns_to_check)

filtered_df.isna().sum()
```

```python
# Get column names of the targets
cols_to_move = ['nutriscore_score', 'nutriscore_grade', 'nova_group']

# Get remaining columns
remaining_cols = [col for col in filtered_df.columns if col not in cols_to_move]

# Reorder
filtered_df = filtered_df[remaining_cols + list(cols_to_move)]
```

```python
import os

directory = "../raw_data"
if not os.path.exists(directory):
    os.makedirs(directory)

output_path = os.path.join(directory, "filtered_df.csv")

filtered_df.to_csv(
    output_path,
    index=False,
    quoting=1,
    quotechar='"',
    escapechar="\\",
    chunksize=1000000
)

# Save it also as a parquet to save space
parquet_output_path = os.path.join(directory, "filtered_df.parquet")
filtered_df.to_parquet(parquet_output_path, index=False)
```
